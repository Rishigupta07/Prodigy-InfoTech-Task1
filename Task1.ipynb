## Bar Plot

## Top ten countries of total population in year 2022 and 2016

```python
# View your data
# print(dc.head())

# Clean column names if needed
dc.columns = dc.columns.str.strip()

# Sort and get top 10 countries by 2022 population
top_10_countries = dc.sort_values(by='2022', ascending=False).head(10)

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 6))

# 2022 plot
plt.subplot(1, 2, 1)
sns.barplot(x="2022", y="Country Code", data=top_10_countries, palette="coolwarm")
plt.title("Top 10 Countries by Population (2022)")
plt.xlabel("Population")
plt.ylabel("Country")

# 2016 plot
plt.subplot(1, 2, 2)
sns.barplot(x="2016", y="Country Code", data=top_10_countries, palette="coolwarm")
plt.title("Top 10 Countries by Population (2016)")
plt.xlabel("Population")
plt.ylabel("Country")

plt.tight_layout()
plt.show()
```

## Extraction of top 10 countries with highest male population

## Top ten countries with heighest male and female poplation in 2022

```python

# Filter for total population
pop_data = dc[dc["Indicator Code"] == "SP.POP.TOTL"].copy()

# Estimate male and female populations
pop_data["Male_2022"] = pop_data["2022"] * 0.505
pop_data["Female_2022"] = pop_data["2022"] * 0.495

# Get top 10 male and female
male_top_ten_countries = pop_data.sort_values(by="Male_2022", ascending=False).head(10)
female_top_ten_countries = pop_data.sort_values(by="Female_2022", ascending=False).head(10)

# Plot
plt.figure(figsize=(16, 8))

# Male bar plot
plt.subplot(1, 2, 1)
sns.barplot(x="Male_2022", y="Country Code", data=male_top_ten_countries, palette="Blues_d")
plt.title("Top 10 Countries by Male Population (2022)", fontsize=12)
plt.xlabel("Male Population", fontsize=10)
plt.ylabel("Country Code", fontsize=10)

# Female bar plot
plt.subplot(1, 2, 2)
sns.barplot(x="Female_2022", y="Country Code", data=female_top_ten_countries, palette="Purples_d")
plt.title("Top 10 Countries by Female Population (2022)", fontsize=12)
plt.xlabel("Female Population", fontsize=10)
plt.ylabel("Country Code", fontsize=10)

plt.tight_layout()
plt.show()
```

## Select the top 10 countries with the highest total population

```python
# Filter total population only
pop_data = dc[dc["Indicator Code"] == "SP.POP.TOTL"].copy()

# Estimate male and female population
pop_data["Male_2022"] = pop_data["2022"] * 0.505
pop_data["Female_2022"] = pop_data["2022"] * 0.495

# Select bottom 10 countries by male and female population
bottom_10_male = pop_data.sort_values(by="Male_2022", ascending=True).head(10)
bottom_10_female = pop_data.sort_values(by="Female_2022", ascending=True).head(10)

# Plot
plt.figure(figsize=(16, 8))

# Male bar plot
plt.subplot(1, 2, 1)
sns.barplot(x="Male_2022", y="Country Code", data=bottom_10_male, palette="Blues_d")
plt.title("Bottom 10 Countries by Male Population (2022)", fontsize=12)
plt.xlabel("Male Population", fontsize=10)
plt.ylabel("Country Code", fontsize=10)

# Female bar plot
plt.subplot(1, 2, 2)
sns.barplot(x="Female_2022", y="Country Code", data=bottom_10_female, palette="Purples_d")
plt.title("Bottom 10 Countries by Female Population (2022)", fontsize=12)
plt.xlabel("Female Population", fontsize=10)
plt.ylabel("Country Code", fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Filter total population only
pop_data = dc[dc["Indicator Code"] == "SP.POP.TOTL"].copy()

# Estimate male and female population
pop_data["Male_2022"] = pop_data["2022"] * 0.505
pop_data["Female_2022"] = pop_data["2022"] * 0.495

# Select bottom 10 countries by total population
bottom_10_countries = pop_data.sort_values(by="2022", ascending=True).head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12, 6))

x = np.arange(len(bottom_10_countries))  # numeric x positions

# Plot female population
plt.bar(x, bottom_10_countries["Female_2022"], color="red", label="Female Population")

# Plot male population stacked on top of female
plt.bar(x, bottom_10_countries["Male_2022"], bottom=bottom_10_countries["Female_2022"], color="green", label="Male Population")

plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Bottom 10 Countries by Total Population (2022) - Stacked by Gender")
plt.xticks(x, bottom_10_countries["Country Code"], rotation=45, ha="right")
plt.legend()
plt.tight_layout()
plt.show()
```
